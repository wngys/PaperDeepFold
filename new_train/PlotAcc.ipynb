```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# chkp = torch.load("/home/wngys/lab/DeepFold/new_model/new_model/model_19.pt")
chkp = torch.load("/home/wngys/lab/DeepFold/new_model/new_model_3/model_9.pt")
print(chkp.keys())
```

```python
# print(chkp['train_acc'])
y1 = []
for t in chkp['train_acc']:
    # print(t[0])
    y1.append(t[0])

x = np.arange(len(chkp['train_acc']))
plt.plot(x, y1)
```

```python
# print(chkp['valid_acc'])
y2 = []
for t in chkp['valid_acc']:
    # print(t[0])
    y2.append(t[0])

x = np.arange(len(chkp['valid_acc']))
plt.plot(x, y2)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
